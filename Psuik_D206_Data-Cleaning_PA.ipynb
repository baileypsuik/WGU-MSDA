The quantitative and qualitative variables were addressed using imputation. Children and Income both showed a right-skewed distribution, so their missing values were imputed using the median (.median()), which is more accurate than the mean in the case of a skewed distribution. Age showed a uniform distribution so missing values were imputed using the forward fill method (.ffill()). This method was used instead of a measure of center (mean or median) because doing so would cause a spike in an otherwise uniform distribution. The forward fill method imputed missing variables based on the row preceding it. Tenure and Bandwidth_GB_Year both showed a bimodal distribution so their missing values were imputed using the mean (.mean()), since this measure of center would help to maintain relative symmetry of the distribution. Qualitative data with missing values was imputed using the mode ((.mode()) of each respective variable. This maintained the mode of the variable by providing a best estimate or probability of what these missing values would be based on the other observations. 

```python
# QUANTITATIVE IMPUTATION
# Children: skewed right -- median
churn_data['Children'].fillna(churn_data['Children'].median(), inplace=True)

# Age: uniform -> forward fill (*imputing with the mean would cause a spike at the mean, and would alter the uniform dist.)
# dataframe.ffill() will propagate last valid observation forward. (Geeks for Geeks, 2023)
# applying ffill() method to fill the missing values 
churn_data['Age'].ffill(axis=0, inplace=True)

# Income: skewed right -- median
churn_data['Income'].fillna(churn_data['Income'].median(), inplace=True)

# Bandwidth: bimodal -- mean
churn_data['Bandwidth_GB_Year'].fillna(churn_data['Bandwidth_GB_Year'].mean(), inplace=True)

# Tenure: bimodal -- mean
churn_data['Tenure'].fillna(churn_data['Tenure'].mean(), inplace=True)
```

```python
# QUALITATIVE IMPUTATION
# Calculating the mode of each categorical variable to use to impute missing values
mode_Techie = churn_data['Techie'].mode()[0]
print(mode_Techie)

mode_IntService = churn_data['InternetService'].mode()[0]
print(mode_IntService)

mode_Phone = churn_data['Phone'].mode()[0]
print(mode_Phone)

mode_TechSupport = churn_data['TechSupport'].mode()[0]
print(mode_TechSupport)
```

```python
# Imputation of above

churn_data['Techie'] = churn_data['Techie'].fillna(mode_Techie)
churn_data['InternetService'] = churn_data['InternetService'].fillna(mode_IntService)
churn_data['Phone'] = churn_data['Phone'].fillna(mode_Phone)
churn_data['TechSupport'] = churn_data['TechSupport'].fillna(mode_TechSupport)
```

```python
# Getting a count of the number of outliers contained in each variable

# Define a function to determine outliers (Kleppen, 2023)
def find_outliers(churn_data):
    q1=churn_data.quantile(0.25)
    q3=churn_data.quantile(0.75)
    IQR=q3-q1
    outliers = churn_data[((churn_data<(q1-1.5*IQR)) | (churn_data>(q3+1.5*IQR)))]
    return outliers

outliers = find_outliers(churn_data[['Lat','Lng','Population','Children','Age','Income','Outage_sec_perweek','Email',
                             'Contacts','Yearly_equip_failure','Tenure','MonthlyCharge','Bandwidth_GB_Year']])

# Print the sum of all non-null values. This sum is the number of outliers for a given variable.
print(outliers.notna().sum())
```

**E1. Principal Components**
<br>To perform principal component analysis (PCA) on the data, all nine quantitative continuous variables in the data set were used. These included: ‘Lng’, ‘Lat’, ‘Income’, ‘Outage_sec_perweek’, ‘Tenure’, ‘MonthlyCharge’, ‘Bandwidth_GB_Year’
<br><br> Below is the code used to create the loadings matrix for the PCA utilizing these variables:

```python
# Import package
from sklearn.decomposition import PCA

# Define features/variables
# Store quantitative continuous variables in a new data frame
churn = churn_data[['Lat','Lng','Income','Outage_sec_perweek','Tenure','MonthlyCharge','Bandwidth_GB_Year']]

# Normalize by using standardization (mean normalization)
# This prevents one variable from being overly influential esp if it is measured in diff units
# Divide mean by standard deviation for each variable. Store in new df
churn_normalized = (churn - churn.mean())/churn.std()
#^ This will create normalized values for all variables in the values within those variables in the dataset
# Create new variable, pca. Output tells you how many principle components you will have
pca = PCA(n_components=churn.shape[1])
print(pca)
```

```python
# Take what you just did and transform it into a dataframe
pca.fit(churn_normalized)

churn_pca = pd.DataFrame(pca.transform(churn_normalized),
                        columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])

# PCA loadings 
# PCs are groupings that have some sort of relationship
# The closer the loading is to 1 or -1 indicates that the variable is significantly associated with the PC
# When assessing which variable is associated with the PC always take the absolute value of the loadings 
# How to get the loading matrix
loadings = pd.DataFrame(pca.components_.T,
                        columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7'],
                        index=churn.columns)
loadings
```

**E2. Criteria Used**
<br>The Kaiser rule recommends retaining principal components (PCs) with eigenvalues greater than or equal to 1 (Middleton, 2023). The eigenvalues represent the strength of the transformation and indicate how meaningful the PC is. The eigenvalues are shown in the scree plot below and indicate that PC1, PC2, PC3, and PC4* should be retained, as they each have eigenvalues greater than or equal to 1. (*Since Python uses zero-indexing, “0” on the scree plot below refers to PC1, “1” refers to PC2, and so on.)
<br><br> A scree plot was created using the following code: 

```python
# A scree plot (simple line segment plot) shows the eigenvalues for each individual PC which helps identify what to keep
# Calculate covariance and vectors then define eigenvalues before creating scree plot
cov_matrix = np.dot(churn_normalized.T, churn_normalized) / churn.shape[0]

eigenvalues = [np.dot(eigenvector.T, np.dot(cov_matrix, eigenvector)) for eigenvector in pca.components_]

plt.plot(eigenvalues)
plt.xlabel('Number of components')
plt.ylabel('eigenvalue')
plt.axhline(y=1, color="red")
plt.show()
```
